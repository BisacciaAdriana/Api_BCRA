```python
api_key='..'
```

```python
import requests
import json
import pandas as pd
import plotly.express as px
url='https://api.estadisticasbcra.com/tasa_prestamos_personales'
headers= {"Content-Type":"application/json","Authorization":f"BEARER {api_key}"}
result= requests.get(url,headers=headers)
print(result.status_code)
```

```python
print(len(result.text))
result.text
```

```python
texto=result.text
json.loads(texto)
```

```python
jsondata=json.loads(texto)
df1=pd.DataFrame.from_dict(jsondata)
df1.head()
```

d que contiene la fecha en formato MySQL (YYYY-MM-DD) y v que contiene el valor.

```python
df1.rename(columns={'d':'Fecha', 'v':'Tasa'},inplace=True)
df1.head()
```

```python
df1.shape
```

```python
url='https://api.estadisticasbcra.com/prestamos'
headers= {"Content-Type":"application/json","Authorization":f"BEARER {api_key}"}
result= requests.get(url,headers=headers)
print(result.status_code)
```

```python
print(len(result.text))
result.text
```

```python
texto=result.text
json.loads(texto)
```

```python
jsondata=json.loads(texto)
df2=pd.DataFrame.from_dict(jsondata)
df2.head()
```

```python
df2.rename(columns={'d':'Fecha', 'v':'Monto'},inplace=True)
df2.head()
```

```python
df2.shape
```

```python
df_merge=pd.merge(df1,df2, on ='Fecha',how= 'inner')
df_merge.head()
```

```python
# esto sirve para que me muestre todo el dataframe completo sin '...', todas las columnas
pd.set_potion('')  
import warnings
warnings.filterwarnings('ignore')
```

```python
df_merge.duplicated()
```

```python
sum(df_merge.isnull().sum())
```

```python
fig=px.line(df_merge, x='Fecha', y='Tasa', title='Tasas de Prestamos Personales')
fig.show()
```

```python
fig=px.line(df_merge, x='Fecha', y='Monto', title='Monto de Prestamos Personales')
fig.show()
```
